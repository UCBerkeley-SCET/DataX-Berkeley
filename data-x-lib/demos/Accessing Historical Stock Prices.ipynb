**Pandas Datareader**

Pandas datateader is an easy-to-use library that provides data sets for various different financial and economic data. Some of the data sources covered by the library include:
* Tiingo
* IEX
* Alpha Vantage
* Econdb
* Enigma
* Quandl
* FRED
* Fama/French
* World Bank
* OECD
* Eurostat
* TSP Fund Data
* Nasdaq Trader Symbol Definitions
* Stooq Index Data
* MOEX Data

The official documentation can be found at https://pandas-datareader.readthedocs.io/en/latest/
